```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import count

plt.style.use('ggplot')

%matplotlib inline
%config InlineBackend.figure_format='svg'
```

```python
def load(path: str) -> pd.DataFrame:
    df = pd.read_csv(path)
    df.sort_values(by='k', ignore_index=True, inplace=True)
    return df
```

```python
df = load('results.csv')
df
```

## Zadanie 2.

```python
def zad2(df):
    print('min: ', df['time'].min())
    print('max: ', df['time'].max())
    print('diff:', df['time'].max() - df['time'].min())
    df.plot(x='k', y='time')
```

```python
zad2(df)
```

## Zadanie 3.

```python
def zad3(df):
    fig, ax = plt.subplots(figsize=(32, 4))
    palette = sns.color_palette('plasma', len(df['rank'].unique()))
    colors = df['rank'].map(dict(zip(count(), palette)))
    df.plot(x='k', y='time', kind='bar', width=1, color=colors, ax=ax)
    ax.set_xticks([])
```

```python
zad3(df)
```

## Zadanie 4.

```python
def zad4(df):
    return df.groupby(by='rank').agg({'k': 'count'})
```

```python
zad4(df)
```

```python
zad4(df).plot(kind='bar')
```

## Zadanie 5.

```python
def zad5(df):
    return df.groupby(by='rank').agg({'time': 'sum'})
```

```python
zad5(df)
```

```python
zad5(df).plot(kind='bar')
```
